```python
#Import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
palette = sns.color_palette()

pd.options.mode.chained_assignment = None

#Import model and pre-processing packages
import xgboost as xgb
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import TimeSeriesSplit

#Change Panadas options so we can see all columns
pd.set_option('display.max_columns', 500)
```

```python
sales_df = pd.read_csv('/Users/benaw/Documents/Personal/Job Applications 2023/.ipynb_checkpoints/Tobacco_Sales.csv', encoding='unicode_escape', parse_dates=True, dayfirst=True, index_col=0)
```

<h2>Feature Manipultaion</h2>

```python
#Rename total sales column
new_cols={'Total':'Total Sales'}

full_df = sales_df.rename(columns=new_cols)
```

```python
#Create features for time series information
full_df['Year'] = full_df.index.year
full_df['Month'] = full_df.index.month
```

```python
#Function for creating a 12 month lag feature for sales
def lags(df, i):
    df['Lag'] = df[i].shift(12)
```

```python
#This is the df we will store our predictions in
predictions = pd.DataFrame(columns=full_df.columns[:-2])

#These are the features we will use for training our models
features = ['Year', 'Month', 'Lag']

"""
The code below will loop through the columns for each province, create a 12 month lag feature and create two dataframes - a 
training df for dates between 2013 and 2016 which will be used for training the model, and a df for 2017 predictions. 
It will then make and train an XGBoost model and then make preditions for 2017; storing them in the 'preditions' dataframe 
defined above.
"""
for i in full_df.columns[:-2]:
    #Create lag feature for the province
    lags(full_df, i)
    
    #Split into training df (before 2017) and prediction df (2017)
    pred_df = full_df[full_df.index >= '2017-01-01']
    train_df = full_df[full_df.index < '2017-01-01']
    
    #Create dfs for training and predictions 
    train_X = train_df[features]
    train_y = train_df[i]

    #Create XGBoost model
    reg = xgb.XGBRegressor(n_estimators=800, 
                           learning_rate=0.01, 
                           booster='gbtree', 
                           early_stopping_rounds=50, 
                           max_depth=3)

    #Fit the model 
    reg.fit(train_X, train_y, 
            eval_set=[(train_X, train_y)], 
            verbose=100)
    
    #Make preditions for 2017
    predictions[i] = reg.predict(pred_df[features])

#Date index for predictions df
predictions.index = pd.date_range('01/01/2017', '12/01/2017', freq='MS')
```

```python
plt.figure(figsize=(15,5))
full_df['Total Sales'].plot()
predictions['Total Sales'].plot()
```

```python
tss = TimeSeriesSplit(n_splits=3, test_size=12)
```

```python
features = ['Year', 'Month', 'Lag']
target = 'Total Sales'

preds = []
scores = []

for train_idx, val_idx in tss.split(full_df):
    #Create df for the fold
    train = full_df.iloc[train_idx]
    val = full_df.iloc[val_idx]
    
    #Create dfs for training and validation 
    train_X = train[features]
    train_y = train[target]
    
    val_X = val[features]
    val_y = val[target]
    
    #Create XGBoost model
    reg = xgb.XGBRegressor(n_estimators=800, 
                           learning_rate=0.01, 
                           booster='gbtree', 
                           early_stopping_rounds=50, 
                           max_depth=3,
                          objective='reg:linear')
    
    #Fit the model to 
    reg.fit(train_X, train_y, eval_set=[(train_X, train_y), (val_X, val_y)], verbose=100)
    
    y_preds = reg.predict(val_X)
    preds.append(y_preds)
    score = np.sqrt(mean_squared_error(val_y, y_preds))
    scores.append(score)
```
