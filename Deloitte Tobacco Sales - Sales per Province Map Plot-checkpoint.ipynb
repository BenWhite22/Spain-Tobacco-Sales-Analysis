```python
import pandas as pd
import geopandas
import matplotlib.pyplot as plt
```

```python
sales_df = pd.read_csv('/Users/benaw/Documents/Personal/Job Applications 2023/.ipynb_checkpoints/Tobacco_Sales.csv', encoding='unicode_escape', parse_dates=True, dayfirst=True, index_col=0)
pop_df = pd.read_csv('/Users/benaw/Documents/Personal/Job Applications 2023/.ipynb_checkpoints/Population.csv', encoding='unicode_escape')

provinces = geopandas.read_file(r"C:\Users\benaw\Downloads\lineas_limite\SHP_ETRS89\recintos_provinciales_inspire_peninbal_etrs89\recintos_provinciales_inspire_peninbal_etrs89.shp")
```

I found a shapefile of Spanish provinces on the Spanish government website and used this to create a geoplot of tobacco sales per year per capita: http://centrodedescargas.cnig.es/CentroDescargas/index.jsp (Límites municipales, provinciales y autonómicos)

```python
#Some of the province names in the SHP file did not match the names in sales_df, so I first had to rename them.
province_rename={'A Coruña':'Coruña (A)', 'Alacant/Alicante':'Alicante / Alacant', 
                 'Araba/Álava':'Álava', 'Bizkaia':'Vizcaya', 'Castelló/Castellón':'Castellón / Castelló',
                 'Gipuzkoa':'Guipúzcoa', 'Illes Balears':'Balears (Illes)',
                 'La Rioja':'Rioja (La)', 'València/Valencia':'Valencia / València'
                }

provinces['NAMEUNIT'] = provinces['NAMEUNIT'].replace(province_rename)

#I also renamed the column title to 'Province' so that I could later merge the dataframes.
provinces = provinces.rename(columns={'NAMEUNIT':'Province'})
```

```python
#Create a dataframe with average sales per year in each province
sales_per_year=pd.DataFrame({'Sales/Year':sales_df.sum()/4, 'Province':sales_df.columns})

#Merge sales_per_year df above with pop_df (which contains the population of each province)
sale_per_cap=pd.merge(sales_per_year, pop_df, on='Province')

#Add a column for sales per year per capita
sale_per_cap['Sales/Year/Capita']=sale_per_cap['Sales/Year']/sale_per_cap['Population (2013)']
```

```python
#Left merge geospatial data and sales data on the 'Province' column
sales_map=pd.merge(provinces, sale_per_cap, on='Province', how='left')
```

```python
#Create the plot
plt.figure(figsize=(15,15))
sales_map.plot(column='Sales/Year/Capita', legend=True)
plt.axis('off')
plt.show()
```

The map above shows that the provinces bordering France have some of the highest tobacco sales per capita. This Statista page (https://www.statista.com/statistics/415034/cigarette-prices-across-europe/) shows that tobacco prices in Spain are significantly lower than in France, so the high sales rates in these provinces is likely caused by people crossing the border to purchase tobacco at a lower price.

Likewise, the Balearic Islands (a popular tourist destination) is the province with the second highest tobacco sales per capita.
