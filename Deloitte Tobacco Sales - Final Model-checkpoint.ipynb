<h5>In this notebook I will create a regression model for predicting tobacco sales in 2017 based on the analysis in 'Deloitte Tobacco Sales - Analysis'</h5>

```python
#Import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd.options.mode.chained_assignment = None

#Import model and pre-processing packages
import xgboost as xgb
from sklearn.metrics import mean_squared_error

#Change Panadas options so we can see all columns
pd.set_option('display.max_columns', 500)
```

```python
sales_df = pd.read_csv('/Users/benaw/Documents/Personal/Job Applications 2023/.ipynb_checkpoints/Tobacco_Sales.csv', encoding='unicode_escape', parse_dates=True, dayfirst=True, index_col=0)
tourism_df = pd.read_csv('/Users/benaw/Documents/Personal/Job Applications 2023/.ipynb_checkpoints/Labour_days_and_Tourism.csv', encoding='unicode_escape', parse_dates=True, dayfirst=True, index_col=0)
```

<h2>Feature Manipultaion</h2>

```python
#Concatenate both dataframes
full_df = pd.concat([sales_df, tourism_df], axis=1)

#Rename total sales column
new_cols={'Total':'Total Sales', 'Num. días laborables': 'Labour Days', 'Pernoctaciones (INE)':'Tourism'}
full_df = full_df.rename(columns=new_cols)
```

```python
#Create features for time series information
full_df['Year'] = full_df.index.year
full_df['Month'] = full_df.index.month

#Create features for tourism and labour lags
full_df['TourismLag12'] = full_df['Tourism'].shift(12)
full_df['TourismLag24'] = full_df['Tourism'].shift(24)
full_df['LabourLag24'] = full_df['Labour Days'].shift(24)
```

```python
full_df
```

```python
#Function for creating 12 and 24 month lag feature for sales
def lags(df, i):
    df['Lag'] = df[i].shift(12)
    df['Lag1'] = df[i].shift(24)
```

```python
#This is the df we will store our predictions in
predictions = pd.DataFrame(columns=full_df.columns[:-5])

#These are the features we will use for training our models
features = ['Year', 'Month', 'Lag', 'Lag1', 'LabourLag24', 'TourismLag12', 'TourismLag24']

"""
The code below will loop through the columns for each province, create 12 month and 24 month lag features and create two 
dataframes - a training df for dates between 2013 and 2016 which will be used for training the model, and a df for 2017 
predictions. It will then make and train an XGBoost model and then make preditions for 2017; storing them in the 'preditions' 
dataframe defined above.
"""
for i in full_df.columns[:-5]:
    #Create lag feature for the province
    lags(full_df, i)
    
    #Split into training df (before 2017) and prediction df (2017)
    pred_df = full_df[full_df.index >= '2017-01-01']
    train_df = full_df[full_df.index < '2017-01-01']
    
    #Create dfs for training features and target feature 
    train_X = train_df[features]
    train_y = train_df[i]

    #Create XGBoost model
    reg = xgb.XGBRegressor(n_estimators=800, 
                           learning_rate=0.01, 
                           booster='gbtree', 
                           early_stopping_rounds=50, 
                           max_depth=3)

    #Fit the model 
    reg.fit(train_X, train_y, 
            eval_set=[(train_X, train_y)], 
            verbose=100)
    
    #Make preditions for 2017
    predictions[i] = reg.predict(pred_df[features])

#Date index for predictions df
predictions.index = pd.date_range('01/01/2017', '12/01/2017', freq='MS')
```

```python
plt.figure(figsize=(12,4))
full_df['Total Sales'].plot(label='Actual Sales')
predictions['Total Sales'].plot(label='Model Predictions')
plt.title('Tobacco Sales 2013-2016 vs 2017 Model Predictions')
plt.ylabel('Tobacco Sales')
plt.legend()
plt.show()
```

```python
#Export to predictions to CSV
predictions.to_csv(r'./.ipynb_checkpoints/2017 Predictions.csv')
```
