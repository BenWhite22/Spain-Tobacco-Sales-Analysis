## 1. Set-up<a class="anchor" id="first-bullet"></a>

```python
#Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
palette = sns.color_palette()

pd.options.mode.chained_assignment = None

#Import model
import xgboost as xgb
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler

pd.set_option('display.max_columns', 500)
```

```python
sales_df = pd.read_csv('/Users/benaw/Documents/Personal/Job Applications 2023/.ipynb_checkpoints/Tobacco_Sales.csv', encoding='unicode_escape', parse_dates=True, dayfirst=True, index_col=0)
tourism_df = pd.read_csv('/Users/benaw/Documents/Personal/Job Applications 2023/.ipynb_checkpoints/Labour_days_and_Tourism.csv', encoding='unicode_escape', parse_dates=True, dayfirst=True, index_col=0)
unem_df = pd.read_csv('/Users/benaw/Documents/Personal/Job Applications 2023/.ipynb_checkpoints/Unemployment.csv', parse_dates=True, dayfirst=True, index_col=0)
prices_df = pd.read_csv('/Users/benaw/Documents/Personal/Job Applications 2023/.ipynb_checkpoints/Prices.csv', parse_dates=True, dayfirst=True, index_col=0)
pop_df = pd.read_csv('/Users/benaw/Documents/Personal/Job Applications 2023/.ipynb_checkpoints/Population.csv', encoding='unicode_escape')
```

```python
#Combine all dataframes
full_df = pd.concat([sales_df, tourism_df, unem_df, prices_df], axis=1)
```

```python
#Rename Columns
new_cols={'Total':'Total Sales', 'Num. días laborables': 'Labour Days', 'Pernoctaciones (INE)':'Tourism','Tasa de paro (INE) ': 'Unemployment', 'Average': 'Average Price'}
full_df = full_df.rename(columns=new_cols)

#Columns for year and month
full_df['Year']=full_df.index.year
full_df['Month']=full_df.index.month
full_df.head(1)
```

```python
#Remove dates from 2017. For analysis we will only use data from 2013-2016.
full_df = full_df[full_df.index < '2017-01-01']
```

## 2. Initial Analysis<a class="anchor" id="second-bullet"></a>

```python
#For analysis I will first create a smaller dataframe containing various features to inspect their correlation with total sales.
columns=['Total Sales', 'Labour Days', 'Tourism', 'Unemployment', 'Average Price', 'Month', 'Year']
data = full_df[columns]
data.head(2)
```

<h3>2.1 Tobacco Sales Over the Full Period</h3>

```python
#12 month rolling average
rolling_avg = data['Total Sales'].rolling(window=12, 
                                           center=True, 
                                           min_periods=6
                                          ).mean()

plt.figure(figsize=(15,5))
data['Total Sales'].plot(kind='line', label='Tobacco Sales')
rolling_avg.plot(label='12 month rolling average')
plt.legend()
plt.ylabel('Tobacco sales')
plt.title('Total tobacco sales across Spain (exl. Canarias) between 2013 and 2016')
plt.show()
```

From the graph above we can see that tobacco sales increase in summer months and decrease in winter, with a slight uptick in December. There has been no overall increase over the period as we can see from the roughly flat 12 month rolling average, and the maximum sales each year has remained fairly consistent.

I am interested to see what caused the sharp increase and then decrease in sales between April and July 2013, and whether the data is erroneous.

```python
sales_df[(sales_df.index>'2013-02-01') & (sales_df.index<'2013-08-01')]
```

By looking at the sales data above I can see that sales increased in nearly every province between March and April 2013, and similarly sales decreased in every region between May and June 2013. This leads me to think that the data is correct and the sales increase/decrease was due to some external factor. I will explore this decrease in relation to other factors such as unemployment, tobacco price, and tourism.

<h3>2.2 Yearly Trend</h3>

Let's see how the variation throughout a year compares year on year. 

```python
#Lag1 = 2013, lag2 = 2014, lag3 = 2015
data1 = data.copy()
data1['lag1'] = data1['Total Sales'].shift(12)
data1['lag2'] = data1['Total Sales'].shift(24)
data1['lag3'] = data1['Total Sales'].shift(36)
```

```python
plt.rcParams.update({'font.size': 7})
fig, ax = plt.subplots(figsize=(10,3))
data1[data1['Year']==2016]['Total Sales'].plot(ax=ax, label='2016')
data1[data1['Year']==2016]['lag1'].plot(ax=ax, label='2015')
data1[data1['Year']==2016]['lag2'].plot(ax=ax, label='2014')
data1[data1['Year']==2016]['lag3'].plot(ax=ax, label='2013')
data1[data1['Year']==2016][['lag1','lag2','lag3','Total Sales']].mean(axis=1).plot(ax=ax, label='Average', style='-.',)
ax.legend()
plt.ylabel('Tobacco sales')
plt.xlabel('Month')
plt.title('Total tobacco sales across Spain (exl. Canarias) between 2013 and 2016, year on year')
plt.show()
```

Again, we can see a clear sales pattern throughout a calendar year. There is a peak every 6 months - a large one in Summer (around July), and a smaller one in Winter (around January).

## 3. Exploring Feature Dependence<a class="anchor" id="third-bullet"></a>

```python
#To explore the correlation between these features, I will first scale the data.
scaler = MinMaxScaler()
columns=['Total Sales', 'Labour Days', 'Tourism', 'Unemployment', 'Average Price', 'Month', 'Year']
df_scaled = pd.DataFrame(scaler.fit_transform(data), columns=data.columns, index=data.index)
```

```python
plt.rcParams.update({'font.size': 7})
#Let's explore the correlation between tourism and tobacco sales.
plt.figure(figsize=(15,3))
df_scaled['Total Sales'].plot(kind='line', label='Tobacco Sales')
df_scaled['Tourism'].plot(label='Tourism')
plt.legend()
plt.title('Total tobacco sales across Spain (exl. Canarias) between 2013 and 2016, against tourism')
plt.show()

#Between tourism and labour days.
plt.figure(figsize=(15,3))
df_scaled['Total Sales'].plot(kind='line', label='Tobacco Sales')
df_scaled['Labour Days'].plot(label='# Labour Days', color=palette[2])
plt.legend()
plt.title('Total tobacco sales across Spain (exl. Canarias) between 2013 and 2016, against num. labour days')
plt.show()

#Between tourism and unemployment.
plt.figure(figsize=(15,3))
df_scaled['Total Sales'].plot(kind='line', label='Tobacco Sales')
df_scaled['Unemployment'].plot(label='Unemployment', color=palette[3])
plt.legend()
plt.title('Total tobacco sales across Spain (exl. Canarias) between 2013 and 2016, against unemployment')
plt.show()

#Between tourism and average tobacco price
plt.figure(figsize=(15,3))
df_scaled['Total Sales'].plot(kind='line', label='Tobacco Sales')
df_scaled['Average Price'].plot(label='Tobacco Price', color=palette[5])
plt.legend()
plt.title('Total tobacco sales across Spain (exl. Canarias) between 2013 and 2016, against tobacco price')
plt.show()
```

There is a strong correlation between the rate of tourism and the seasonal trend of tobacco sales, i.e. both tourism and sales increase in summer, and decrease in winter. However, this might not mean causation - the seasonal increase could instead be caused by temperature.<br><br>
There is a slight correlation between labour days and tobacco sales - tobacco sales decrease when there are less available days to purchase tobacco, as expected. <br><br>
There is no correlation between unemployment and tobacco price and tobacco sales.

```python
#Heatmap of the correlations
columns = ['Total Sales', 'Labour Days', 'Tourism', 'Unemployment', 'Average Price']
plt.figure(figsize=(5,5))
hm = sns.heatmap(df_scaled[columns].corr(), annot=True, fmt="0.2f")
hm.set_title('Correlation between features')
plt.show()
```

<h3>3.1 Serial Dependence</h3>

```python
#Now I will plot an autocorrelation plot to see if there is any serial dependence.
plt.figure(figsize=(8,3))
pd.plotting.autocorrelation_plot(data['Total Sales'])
plt.axvline(x = 6, color = 'r', linestyle=':')
plt.axvline(x = 12, color = 'r', linestyle=':')
plt.axvline(x = 18, color = 'r', linestyle=':')
plt.axvline(x = 24, color = 'r', linestyle=':')
plt.title('Autocorrelation plot showing serial dependence on time lags')
plt.show()
```

The solid grey horizontal line represents the 95% confidence interval, and the dotted grey line represents the 99% confidence interval.

The lags which breach the 99% CI (seen in red) are multiples of 6 (6, 12, 18, 24 months), which makes sense given that we have already identified an annual pattern of tobacco sales, with peaks every 6 months.  We can also see peaks at 24 months and 36 months, however the autocorrelation for these lags 

```python
#Create 6, 12, 18, and 24 month lag features, and view the correlation.
data['Lag6'] = data['Total Sales'].shift(6)
data['Lag12'] = data['Total Sales'].shift(12)
data['Lag18'] = data['Total Sales'].shift(18)
data['Lag24'] = data['Total Sales'].shift(24)
hm = sns.heatmap(data[['Total Sales', 'Lag6', 'Lag12', 'Lag18', 'Lag24']].corr(), annot=True, fmt="0.2f")
```

## 4. Building a Preliminary Model<a class="anchor" id="fourth-bullet"></a>

```python
#Define training features
features = ['Labour Days','Tourism','Unemployment','Average Price','Month', 'Year','Lag6','Lag12','Lag18','Lag24']

#We'll use 01/01/2013-01/12/2015 for training, and 2016 for validation.
train_df = data[data.index < '2016-01-01']
val_df = data[data.index >= '2016-01-01']

train_X = train_df[features]
train_y = train_df['Total Sales']

val_X = val_df[features]
val_y = val_df['Total Sales']
```

```python
#This DF will store our predictions
predictions = pd.DataFrame(val_df['Total Sales'], index=val_df.index)

#Create XGBoost model
reg = xgb.XGBRegressor(n_estimators=1000, 
                       learning_rate=0.01, 
                       booster='gbtree', 
                       early_stopping_rounds=50, 
                       max_depth=3)

#Fit the model 
reg.fit(train_X, train_y, 
        eval_set=[(train_X, train_y)], 
        verbose=100)

#Make predictions and store in predictions df
pred = reg.predict(val_X)
predictions['Preds'] = pred

#Assess predictions using root mean squared error
score = np.sqrt(mean_squared_error(val_y, pred))
score
```

<h3>4.1 Assessing the Model</h3>

```python
#Visualising our 2016 predictions against actual sales
plt.figure(figsize=(15,5))
predictions['Total Sales'].plot(color='r', label='Actual Sales')
predictions['Preds'].plot(color='b', label='Prediction')
plt.xlabel('Date')
plt.ylabel('Total Tobacco Sales')
plt.title('Total Tobacco Sales in Spain (Exl. Canaries) in 2016 vs Regression Model Predictions')
plt.legend()
plt.show()
```

```python
#Let's plot feature importance to see how the model weighs each feature
feat_import = reg.feature_importances_
imp = pd.DataFrame({'Feature':features, 'Importance':feat_import})
imp.plot(kind='bar', x='Feature')
plt.show()
```

```python
new_pred = reg.predict(val_X)
predictions['Price Increased Preds'] = new_pred
```
