## 1. Set-up<a class="anchor" id="first-bullet"></a>

## 2. Initial Analysis<a class="anchor" id="second-bullet"></a>

```python
#For analysis I will first create a smaller dataframe containing various features to inspect their correlation with total sales.
columns=['Total Sales', 'Labour Days', 'Tourism', 'Unemployment', 'Average Price', 'Month', 'Year']
data = full_df[columns]
data.head(2)
```

<h3>2.1 Tobacco Sales Over the Full Period</h3>

```python
#12 month rolling average
rolling_avg = data['Total Sales'].rolling(window=12, 
                                           center=True, 
                                           min_periods=6
                                          ).mean()

plt.figure(figsize=(15,5))
data['Total Sales'].plot(kind='line', label='Tobacco Sales')
rolling_avg.plot(label='12 month rolling average')
plt.legend()
plt.ylabel('Tobacco sales')
plt.title('Total tobacco sales across Spain (exl. Canarias) between 2013 and 2016')
plt.show()
```

From the graph above we can see that tobacco sales increase in summer months and decrease in winter, with a slight uptick in December. There has been no overall increase over the period as we can see from the roughly flat 12 month rolling average, and the maximum sales each year has remained fairly consistent.

<h3>2.2 Yearly Trend</h3>

Let's see how the variation throughout a year compares year on year. 

```python
#Lag1 = 2013, lag2 = 2014, lag3 = 2015
data1 = data.copy()
data1['lag1'] = data1['Total Sales'].shift(12)
data1['lag2'] = data1['Total Sales'].shift(24)
data1['lag3'] = data1['Total Sales'].shift(36)
```

```python
plt.rcParams.update({'font.size': 7})
fig, ax = plt.subplots(figsize=(10,3))
data1[data1['Year']==2016]['Total Sales'].plot(ax=ax, label='2016')
data1[data1['Year']==2016]['lag1'].plot(ax=ax, label='2015')
data1[data1['Year']==2016]['lag2'].plot(ax=ax, label='2014')
data1[data1['Year']==2016]['lag3'].plot(ax=ax, label='2013')
data1[data1['Year']==2016][['lag1','lag2','lag3','Total Sales']].mean(axis=1).plot(ax=ax, label='Average', style='-.',)
ax.legend()
plt.ylabel('Tobacco sales')
plt.xlabel('Month')
plt.title('Total tobacco sales across Spain (exl. Canarias) between 2013 and 2016, year on year')
plt.show()
```

Again, we can see a clear sales pattern throughout a calendar year. There is a peak every 6 months - a large one in Summer (around July), and a smaller one in Winter (around December).

```python
#Heatmap of the correlations
columns = ['Total Sales', 'Labour Days', 'Tourism', 'Unemployment', 'Average Price']
plt.figure(figsize=(5,5))
hm = sns.heatmap(df_scaled[columns].corr(), annot=True, fmt="0.2f")
hm.set_title('Correlation between features')
plt.show()
```

<h5>Tourism</h5>
There is a strong correlation between the rate of tourism and the annual trend of tobacco sales, i.e. both tourism and sales increase in summer, and decrease in winter. However, this might not mean causation - the seasonal increase could instead be caused by temperature.<br>
<h5>Labour Days</h5>
There is a strong correlation between labour days and sales - tobacco sales decrease when there are less available days to purchase tobacco, as expected. <br>
<h5>Unemployment</h5>
There is a slight negative correlation between unemployment and sales, i.e. sales increase as unemployment decreases and people have more disposable income. 
<h5>Tobacco Price</h5>
Surprisingly, according to the dataset we have, there is no correlation between sales and tobacco price. I think there are 3 possible explanations for this, which I would be interested to explore in more detail if I had time:<br>
1. Inaccurate/faulty data: The El Precio del Tabaco website may not be accurate - I would like to compare the data against another website to check its accuracy.<br>
2. Inflation: The prices on the source website may not be adjusted for inflation.<br>
3. The time period is too small: In the sales data we observe a yearly pattern whereby tobacco sales increase in summer and decline in winter, however over any given year tobacco prices remain relatively stable, i.e. there is no seasonal pattern to prices. I suspect that price will correlate more strongly with net increases/decreases of sales, rather than with seasonal variations. The first graph in section 2.1 shows there is no overall increase/decrease in sales between 2013 and 2016, however this is a small timeframe in the grand scheme of things. If we zoomed out and observed tobacco sales over say 100 years, we will note a net increase/decrease, which will likely correlate with tobacco price.<br>

<h3>3.1 Serial Dependence</h3>

```python
#Now I will plot an autocorrelation plot to see if there is any serial dependence.
plt.figure(figsize=(8,3))
pd.plotting.autocorrelation_plot(data['Total Sales'])
plt.axvline(x = 6, color = 'r', linestyle=':')
plt.axvline(x = 12, color = 'r', linestyle=':')
plt.axvline(x = 18, color = 'r', linestyle=':')
plt.axvline(x = 24, color = 'r', linestyle=':')
plt.title('Autocorrelation plot showing serial dependence on time lags')
plt.show()
```

The solid grey horizontal line represents the 95% confidence interval, and the dotted grey line represents the 99% confidence interval.

The lags which breach the 99% CI (seen in red) are multiples of 6 (6, 12, 18, 24 months), which makes sense given that we have already identified an annual pattern of tobacco sales, with peaks every 6 months.  We can also see peaks at 30 months and 36 months, however the autocorrelation for these lags is too small, so I will not include them in my model.

```python
cols = ['Total Sales', 'Lag6', 'Lag12', 'Lag18', 'Lag24', 
                       'TourismLag12', 'TourismLag24', 'LabourLag12','LabourLag24']
vec = pd.DataFrame({'Total Sales':data[cols].corr()['Total Sales'][1:]})
vec
```

```python
#Create 6, 12, 18, and 24 month sales lag features and lag features for labour days and toursism, then view the correlation.
data['Lag6'] = data['Total Sales'].shift(6)
data['Lag12'] = data['Total Sales'].shift(12)
data['Lag18'] = data['Total Sales'].shift(18)
data['Lag24'] = data['Total Sales'].shift(24)
data['TourismLag12'] = data['Tourism'].shift(12)
data['TourismLag24'] = data['Tourism'].shift(24)
data['LabourLag12'] = data['Labour Days'].shift(12)
data['LabourLag24'] = data['Labour Days'].shift(24)
sns.heatmap(vec.sort_values(by='Total Sales', ascending=False), annot=True, fmt="0.2f", linewidth=1)
plt.title('Correlation Between Lag Features and Tobacco Sales')
plt.show()
```

## 4. Building a Preliminary Model<a class="anchor" id="fourth-bullet"></a>

```python
#Define training features
features = ['Labour Days','Tourism','Unemployment','Average Price','Month', 'Year', 
            'Lag6', 'Lag12', 'Lag18', 'Lag24', 
            'TourismLag12', 'TourismLag24', 
            'LabourLag12','LabourLag24']

#We'll use 01/01/2013-01/12/2015 for training, and 2016 for validation.
train_df = data[data.index < '2016-01-01']
val_df = data[data.index >= '2016-01-01']

train_X = train_df[features]
train_y = train_df['Total Sales']

val_X = val_df[features]
val_y = val_df['Total Sales']
```

```python
#This DF will store our predictions
predictions = pd.DataFrame(val_df['Total Sales'], index=val_df.index)

#Create XGBoost model
reg = xgb.XGBRegressor(n_estimators=1000, 
                       learning_rate=0.01, 
                       booster='gbtree', 
                       early_stopping_rounds=50, 
                       max_depth=3)

#Fit the model 
reg.fit(train_X, train_y, 
        eval_set=[(train_X, train_y)], 
        verbose=100)

#Make predictions and store in predictions df
pred = reg.predict(val_X)
predictions['Preds'] = pred

#Assess predictions using root mean squared error
score = np.sqrt(mean_squared_error(val_y, pred))
score
```

```python
#What % does the rmse represent?
score/val_y.mean()*100
```

The model predicted sales in 2016 with 94.68% accuracy.

<h3>4.1 Assessing the Model</h3>

```python
#Visualising our 2016 predictions against actual sales
plt.rcParams.update({'font.size': 10})
plt.figure(figsize=(12,4))
predictions['Total Sales'].plot(color='r', label='Actual Sales')
predictions['Preds'].plot(color='b', label='Prediction')
plt.xlabel('Date')
plt.ylabel('Total Tobacco Sales')
plt.title('Total Tobacco Sales in Spain (Exl. Canaries) in 2016 vs Regression Model Predictions')
plt.legend()
plt.show()
```

```python
#Let's plot feature importance to see how the model weighs each feature
plt.figure(figsize=(12,4))
feat_import = reg.feature_importances_
imp = pd.DataFrame({'Feature':features, 'Importance':feat_import})
imp.sort_values(by='Importance', ascending=True).plot(kind='barh', x='Feature')
plt.title('Weighting Assigned to Features by XGBoost Model')
plt.show()
```

<h3>4.2 What effect would a 10% price increase have on tobacco sales?</h3>

Now that we have a model for predicting tobacco sales, we can pass in a new val_X dataframe with the average tobacco price increased by 10% (all other features will remain the same).

```python
val_X['Average Price'] = val_X['Average Price']*1.1
val_X
```

```python
#2016 predictions with a 10% price increase
new_pred = reg.predict(val_X)
predictions['Price Increased Preds'] = new_pred
```

```python
#Visualising our price-increased predictions against previous predictions against actual sales
plt.figure(figsize=(12,4))
predictions['Total Sales'].plot(color='r', label='Actual Sales')
predictions['Preds'].plot(color='b', label='Prediction')
predictions['Price Increased Preds'].plot(color='g', label='10% Price-Increased Prediction')
plt.xlabel('Date')
plt.ylabel('Total Tobacco Sales')
plt.title('Total Tobacco Sales in Spain (Exl. Canaries) in 2016 vs Regression Model Predictions')
plt.legend()
plt.show()
```

```python
#Numerically comparing price-increased predictions against previous predictions
predictions['Price Increased Preds']-predictions['Preds']
```

Counter-intuitively, the 10% price increase made no difference to our predictions. However, this was to be expected given the null correlation between sales and price, as discussed in section 3. Please see section 3 for possible explanations for this.
